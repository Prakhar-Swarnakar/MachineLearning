```python
import matplotlib.pyplot as plt
```

```python
line_data = [
    (1, 2.79),
    (2, 6.42),
    (3, 8.31),
    (4, 11.32),
    (5, 14.70),
    (6, 18.08),
    (7, 20.68),
    (8, 25.32),
    (9, 28.64),
    (10, 31.95)
]
```

```python
x_line = [line_data[x][0] for x in range(len(line_data))]
x_line
y_line = [line_data[x][1] for x in range(len(line_data))]
y_line
```

```python
plt.plot(x_line, y_line, marker='o', linestyle='-', color='red', label='Data Line')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot Example')
plt.grid(True)
plt.legend()
```

```python
plt.show()
```

```python
plt.scatter(x_line, y_line, marker='+', color='b', s= 100,label='Data Points')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values1 = [15, 24, 12, 30]
values2 = [10, 18, 8, 25]

# Basic bar plot
plt.figure(figsize=(8, 6))
plt.bar(categories, values1, color='blue', label='Value 1')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Basic Bar Plot')
plt.legend()
plt.show()

# Stacked bar plot
plt.figure(figsize=(8, 6))
plt.bar(categories, values1, color='blue', label='Value 1')
plt.bar(categories, values2, color='orange', bottom=values1, label='Value 2')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Plot')
plt.legend()
plt.show()

# Grouped bar plot
bar_width = 0.35
index = np.arange(len(categories))
plt.figure(figsize=(8, 6))
plt.bar(index, values1, width=bar_width, color='blue', label='Value 1')
plt.bar(index + bar_width, values2, width=bar_width, color='orange', label='Value 2')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Grouped Bar Plot')
plt.xticks(index + bar_width / 2, categories)
plt.legend()
plt.show()

# Horizontal bar plot
plt.figure(figsize=(8, 6))
plt.barh(categories, values1, color='green', label='Value 1')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Plot')
plt.legend()
plt.show()
```

Basic Bar Plot: A simple bar plot with default coloring.
Stacked Bar Plot: Two datasets stacked on top of each other.
Grouped Bar Plot: Two datasets grouped side by side.
Horizontal Bar Plot: Bars are displayed horizontally.
